### Задание 2

```python
t = np.linspace(0, 10, num=51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color= 'green')
plt.axis([0.5, 9.5,-2.5, 2.5])
plt.xlim([0.5, 9.5])
plt.ylim([-2.5, 2.5])
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.show()
```

### Задание 3

```python
x = np.linspace(-3, 3, num=51)
x
```

```python
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows= 2, ncols= 2)
[График1, График2, График3, График4] = ax.flatten()

График1.plot(x, y1)
График2.plot(x, y2)
График3.plot(x, y3)
График4.plot(x, y4)

График1.axis(xmin=-5,xmax=5)
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3)
```
